```python
import findspark
findspark.init()
```

```python
from pyspark.context import SparkContext
from pyspark.sql import SparkSession
```

```python
sc = SparkContext.getOrCreate()
spark = SparkSession(sc)
```

```python
df_stock = spark.read \
    .option('header', True) \
    .option('inferSchema', True) \
    .csv(path='data/appl_stock.csv')

df_stock.show()
```

```python
df_stock.printSchema()
```

```python
df_stock.head(3)[0]
```

```python
df_stock.head(1)
```

```python
df_stock.filter('Close < 500').show()
```

```python
df_stock.filter('Close < 500').select(['open', 'close']).show()
```

```python
df_stock.filter(df_stock['Close'] < 500).select(['open', 'close', 'volume']).show()
```

```python
df_stock.filter( (df_stock['Close'] < 200) & (df_stock['Open']>200) ).show()
```

```python
df_stock.filter( (df_stock['Close'] < 200) & ~(df_stock['open']>200) ).show()
```

```python
results = df_stock.filter(df_stock['Close']<200).collect()
results
```
