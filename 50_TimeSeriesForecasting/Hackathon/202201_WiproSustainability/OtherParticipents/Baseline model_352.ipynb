```python
from sklearn.datasets import make_multilabel_classification
# this will generate a random multi-label dataset
X, y = make_multilabel_classification(sparse = True, n_labels = 20,
return_indicator = 'sparse', allow_unlabeled = False)
```

```python
#removinf these columns bcz their correlation is very less
train = df_train.drop(['Year','Month','Day'],axis=1)
```

```python
X=train.drop(['Clearsky DHI','Clearsky DNI','Clearsky GHI'],axis=1)
y=train[['Clearsky DHI','Clearsky DNI','Clearsky GHI']]
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
X[['Temperature','Pressure',
       'Relative Humidity', 'Solar Zenith Angle', 'Precipitable Water',
       'Wind Direction', 'Wind Speed']]=scalar.fit_transform(X[['Temperature','Pressure',
       'Relative Humidity', 'Solar Zenith Angle', 'Precipitable Water',
       'Wind Direction', 'Wind Speed']])
```

```python
#scaled features
X.head()
```

```python
y.head()
```

```python
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.multioutput import MultiOutputRegressor
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, train_size=400, test_size=200, random_state=4
)
max_depth = 30
regr_multirf = MultiOutputRegressor(
    RandomForestRegressor(n_estimators=100, max_depth=max_depth, random_state=0)
)
regr_multirf.fit(X_train, y_train)
```

```python
y_m = regr_multirf.predict(X_test)
```

```python
regr_rf = RandomForestRegressor(n_estimators=100, max_depth=max_depth, random_state=2)
regr_rf.fit(X_train, y_train)
# Predict on new data
y_rf = regr_rf.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
print("RMSE of multi output regresor",mean_squared_error(y_test, y_m,squared=False))
print("RMSE of RF regresor",mean_squared_error(y_test, y_rf,squared=False))
```

```python
test_data = pd.read_csv('test.csv')
```

```python
def process_data(df):
    X=df.drop(['Year','Month','Day'],axis=1)
    scalar = StandardScaler()
    X[['Temperature','Pressure',
       'Relative Humidity', 'Solar Zenith Angle', 'Precipitable Water',
       'Wind Direction', 'Wind Speed']]=scalar.fit_transform(X[['Temperature','Pressure',
       'Relative Humidity', 'Solar Zenith Angle', 'Precipitable Water',
       'Wind Direction', 'Wind Speed']])
    y=X.drop(['Clearsky DHI','Clearsky DNI','Clearsky GHI'],axis=1)
    return y
```

```python
test = process_data(test_data)
```
