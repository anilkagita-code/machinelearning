```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_house = pd.read_csv('data/housingprice.csv')
df_house.sample(5)
```

```python
df_house.shape
```

```python
pd.crosstab(df_house['MSSubClass'], df_house['Street'])
```

```python
pd.crosstab(df_house['MSSubClass'], df_house['Street'], margins=True)
```

```python
pd.crosstab(df_house['MSSubClass'], df_house['Street'], margins=True, normalize=True)
```

```python
# The data is normalized for row wise
pd.crosstab(df_house['MSSubClass'], df_house['Street'], margins=True, normalize='index')
```

```python
# The data is normalized column wise
pd.crosstab(df_house['MSSubClass'], df_house['Street'], margins=True, normalize='columns')
```

```python
pd.crosstab(df_house['MSSubClass'], df_house['Street'], values=df_house['LotArea'], aggfunc='mean')
```
