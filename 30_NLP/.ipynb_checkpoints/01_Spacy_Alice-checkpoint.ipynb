```python
text = '''"Let's go to N.Y.!"'''
```

```python
from beautifultable import BeautifulTable
```

```python
table = BeautifulTable()
```

```python
table.columns.header = ['Text', 'POS', 'DEP', 'LEMMA', 'TAG', 'NORM']
doc = nlp(text)
for token in doc:
    table.rows.append([token.text, token.pos_,  token.dep_, token.lemma_, token.tag_,  token.norm_])
    
print(table)
```

```python
text = 'Apple is looking at buying UK startup for $1 billion.'
doc = nlp(text)

table = BeautifulTable()

table.columns.header = ['Text', 'POS', 'DEP', 'LEMMA', 'TAG', 'NORM']

for token in doc:
    table.append_row([token.text, token.pos_, token.dep_, token.lemma_, token.tag_, token.norm_])
    
    
print(table)
```
