```python
X = df_house.drop(columns='price', axis=1)
y = df_house['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mmscaler = MinMaxScaler(feature_range=(0, 1))
```

```python
X_train_sc = mmscaler.fit_transform(X_train)
X_test_sc = mmscaler.transform(X_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Input
```

```python
X_train_sc.shape
```

```python
i0 = Input(shape=X_train_sc.shape[1])

h0 = Dense(19, activation=tf.nn.relu)(i0)
h0 = Dense(19, activation=tf.nn.relu)(h0)
h0 = Dense(19, activation=tf.nn.relu)(h0)
h0 = Dense(19, activation=tf.nn.relu)(h0)

y = Dense(1)(h0)
mod_basic = Model(inputs=i0, outputs=y)
```

```python
mod_basic.summary()
```

```python
mod_basic.compile(optimizer=tf.keras.optimizers.Adam(), 
                 loss=tf.keras.losses.MeanSquaredError())
```

```python
mod_basic.fit(x=X_train_sc, y=y_train, 
              validation_data=(X_test_sc, y_test), batch_size=128,
             epochs=400)
```

```python
pd.DataFrame(mod_basic.history.history).plot()
```
