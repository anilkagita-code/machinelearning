```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_pieces = pd.read_csv('data/Piece_Dim.csv')
df_pieces.sample(5)
```

## Histogram

```python
sns.histplot(data=df_pieces, x='Length', bins=10)
plt.tight_layout()
plt.show()
```

```python
sns.displot(data=df_pieces, kind='hist', x='Length', bins=10)
plt.show()
```

```python
sns.displot(data=df_pieces, kind='hist', x='Length', bins=10, hue='Operator')
plt.show()
```

```python
sns.displot(data=df_pieces, kind='hist', x='Length', bins=10, col='Operator')
plt.show()
```

```python
sns.displot(data=df_pieces, kind='kde', x='Length')
plt.show()
```

```python
sns.displot(data=df_pieces, kind='kde', x='Length', hue='Operator')
plt.show()
```

```python
sns.displot(data=df_pieces, kind='kde', x='Length', hue='Operator', rug=True)
plt.show()
```

```python
sns.displot(data=df_pieces, kind='ecdf', x='Length', hue='Operator', rug=True)
plt.show()
```

```python
sns.displot(data=df_pieces, kind='hist', x='Length', kde=True, rug=True)
plt.show()
```

```python
sns.displot(data=df_pieces, kind='hist', x='Length', hue='Operator', kde=True, rug=True)
plt.show()
```

```python
sns.displot(data=df_pieces, kind='hist', x='Length', col='Operator', kde=True, rug=True)
plt.show()
```

```python
sns.displot(data=df_pieces, x='Length', y='Width')
```

```python
sns.displot(data=df_pieces, x='Length', y='Width', kind='kde')
plt.show()
```

```python
sns.displot(data=df_pieces, x='Length', y='Width', kind='kde', hue='Operator')
plt.show()
```

## Category Plots

Categorical scatterplots:

    stripplot() (with kind="strip"; the default)

    swarmplot() (with kind="swarm")

Categorical distribution plots:

    boxplot() (with kind="box")

    violinplot() (with kind="violin")

    boxenplot() (with kind="boxen")

Categorical estimate plots:

    pointplot() (with kind="point")

    barplot() (with kind="bar")

    countplot() (with kind="count")


```python
sns.catplot(data=df_pieces, x='Operator', y='Length')
```

```python
sns.catplot(data=df_pieces, x='Operator', y='Length', kind='strip')
```

```python
sns.catplot(data=df_pieces, x='Operator', y='Length', kind='swarm')
```

```python
sns.catplot(data=df_pieces, x='Operator', y='Length', kind='box')
```

```python
sns.catplot(data=df_pieces, x='Length', y='Operator', kind='box')
```

```python
sns.catplot(data=df_pieces, col='Operator', y='Length', kind='violin' )
```

```python
sns.catplot(data=df_pieces, x='Operator', y='Length', kind='violin' )
```

```python
sns.catplot(data=df_pieces, x='Operator', y='Length', kind='boxen')
```

```python
# This gives the mean of each operator and 95% confidence interval
sns.catplot(data=df_pieces, x='Operator', y='Length', kind='point')
```

```python
sns.catplot(data=df_pieces, x='Operator', y='Length', kind='point')
sns.catplot(data=df_pieces, x='Operator', y='Length', kind='bar')
plt.show()
```

```python
sns.catplot(data=df_pieces, x="Operator", kind='count')
```

## Rel Plots

```python
sns.relplot(data=df_pieces, x='Length', y='Width')
```

```python
sns.relplot(data=df_pieces, x='Length', y='Width', kind='scatter', hue='Operator')
```

```python
sns.relplot(data=df_pieces, x='Length', y='Width', kind='scatter', col='Operator')
```

```python
sns.relplot(data=df_pieces, x='Length', y='Width', kind='scatter', row='Operator')
```

```python
sns.relplot(data=df_pieces, x='Item_No', y='Length', kind='line')
```

```python
sns.relplot(data=df_pieces, x='Item_No', y='Length', kind='line', aspect=3)
```

```python
sns.relplot(data=df_pieces, x='Item_No', y='Length', kind='line', aspect=3)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.relplot(data=df_pieces, x='Item_No', y='Length', kind='line', aspect=3, hue='Operator')
plt.xticks(rotation=90)
plt.show()
```

```python
# Regression between Length and Width
sns.regplot(data=df_pieces, x='Length', y='Width')
```
