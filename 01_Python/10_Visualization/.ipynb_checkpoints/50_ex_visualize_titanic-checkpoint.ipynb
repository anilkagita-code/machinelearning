```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
df_titanic = sns.load_dataset(name='titanic')
```

```python
df_titanic.sample(10)
```

```python
df_titanic.describe()
```

```python
df_titanic.info()
```

```python
sns.displot(data=df_titanic, x='age', kde=True)
```

```python
sns.displot(data=df_titanic, x='age', hue='sex', kde=True)
```

```python
sns.displot(data=df_titanic, x='age', col='sex', kde=True)
```

```python
sns.displot(data=df_titanic, x='age', hue='sex', kde=True, multiple='dodge')
```

```python
sns.displot(data=df_titanic, x='age', hue='sex', kde=True, multiple='dodge')
plt.tight_layout()
plt.show()
```

```python
sns.displot(data=df_titanic, x='age', col='sex', hue='survived', kde=True, multiple='dodge')
plt.tight_layout()
plt.show()
```

```python
sns.displot(data=df_titanic, x='age', col='sex', hue='survived', kde=True, multiple='dodge', bins=8)
plt.tight_layout()
plt.show()
```

```python
sns.displot(data=df_titanic, x='age', col='sex', row='survived', kde=True, multiple='dodge', bins=8)
plt.tight_layout()
plt.show()
```

```python
sns.catplot(data=df_titanic, x='survived', y='age', hue='sex')
```

```python
sns.catplot(data=df_titanic, x='survived', y='age', hue='sex', kind='swarm', aspect=2)
```

```python
sns.catplot(data=df_titanic, x='survived', y='age', hue='sex', kind='violin', aspect=2)
```

```python
sns.catplot(data=df_titanic, x='survived', y='age', hue='sex', kind='box', aspect=2)
```

```python
sns.catplot(data=df_titanic, x='sex', kind='count', hue='survived')
```

```python
sns.catplot(data=df_titanic, x='deck', kind='count', hue='survived')
```
