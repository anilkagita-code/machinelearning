```python
filt = (sur_df['Country']=='India')
sur_df[filt]
```

```python
sur_df.loc[~((sur_df['Country']=='India') & (sur_df['LanguageDesireNextYear']=='Python') \
       & ((sur_df['Hobbyist']=='Yes') | (sur_df['Student']=='Yes')))]
```
